## Problem 1 

A drawer contains red socks and black socks. When two socks are drawn at random, the probability that both are red is $0.5$. 

- How small can the number of socks in the drawer be? 

- How small if the number of black socks is even? 

Total points: [6]

### Solution Submission

#### Math 1

The probability of choosing two red socks in a row at random can be described as:

$$\frac{total\ number\ of\ ways\ to\ choose\ 2\ red\ socks}{ total\ number\ of\ ways\ to\ pick\ any\ 2\ socks}$$

That is, we have:

$
\begin{align}
p(r = 2) = \frac{\pmatrix{r \\ 2}}{\pmatrix{n \\ 2}} =
    \frac{\frac{r!}{2!(r-2)!}}{\frac{n!}{2!(n-2!}} = 
    \frac{r(r-1)}{n(n-1)}
\end{align}
$

Since $p(r = 2) = 0.5$, and $n = r + b$, then we can solve for r in terms of b:

$
\begin{align*} \label{eq2}
p(r=2) & \Rightarrow \frac{r(r-1)}{n(n-1)} = \frac{r(r-1)}{(r+b)(r+b-1)} = 0.5 \\
& \Rightarrow r(r-1) = 0.5(r+b)(r+b-1) \\
& \Rightarrow r^2 - r = 0.5[r^2 + b^2 + 2rb -r -b] \\
& \Rightarrow r^2(0.5) + r(-0.5 - b) + 0.5b(1 - b) = 0 \\
& \Rightarrow r = \frac{-(-0.5 - b) \pm \sqrt{(-0.5-b)^2 - 4(0.5)(0.5b)(1-b)}}{2(0.5)} \\
& \Rightarrow r = b + 0.5 \pm \sqrt{b^2 + 0.25 + b -b(1-b)} \\ \\
& \Rightarrow r = b + 0.5 \pm \sqrt{2b^2 + 0.25}
\end{align*}
$


Note that we will only use the addition of the $\sqrt{2b^2 + 0.25}$ since $r>b$ and $r,b \in \mathbb{Z} > 0$ must hold for $p(r=2)=0.5$.

#### Code 1

(i) Using probability generating functions, show that if $X \sim Binomial \left(n, p \right)$ and $Y \sim Binomial \left(m, p\right)$, then $Z \sim Binomial \left(n+m, p \right)$. [4]

### Solution Submission

#### Part 2-i

##### Math 2-i

We have that for two independent random variables $X$ and $Y$ that $S = X + Y$ [1]. Then:

$
\begin{align*}
P_S(s) = \sum_{x\in \mathcal{X}} P_{X}(x)P_{Y}(s-x)
\end{align*}
$

Therefore, given a Binomial distribution has a probability mass(density) function of the form [2]:

$
\begin{align*}
P_{Y\sim B(m,p)}(y) = \pmatrix{m \\ y} p^y (1-p)^{m-y},\ \text{for}\ y\in \{0,1,...,m\},\ 0 \le p \le 1
\end{align*}
$

We then have that $P_Z(z)$ for $Z = X+Y$ is:

$
\begin{align*}
P_{Z}(z) &= \sum_{x\in \mathcal{X}} P_{X}(x)P_{Y}(z-x) \\
&= \sum_{x\in \mathcal{X}} \pmatrix{n \\ x} p^x (1-p)^{n-x} \pmatrix{m \\ z-x} p^{z-x} (1-p)^{m-(z-x)} \\
&= \sum_{x\in \mathcal{X}} \pmatrix{n \\ x} \pmatrix{m \\ z-x} p^{z} (1-p)^{n+m-z} \\
&= p^{z} (1-p)^{n+m-z} \sum_{x\in \mathcal{X}} \pmatrix{n \\ x} \pmatrix{m \\ z-x} \Rightarrow \text{MATLAB symbolic evaluation} \\
&= p^{z} (1-p)^{n+m-z} \frac{\Gamma(m+n+1)}{\Gamma(z+1)\Gamma(m+n-z+1)} \\ \\
&= p^{z} (1-p)^{n+m-z} \pmatrix{n+m \\ z}
\end{align*}
$

We will additionally show numerically that summation perfomed by MATLAB is equivalent to $\pmatrix{n+m \\ z}$ by picking numbers from the $X$ and $Y$ distributions ***and plotting below***

That is, we should end up with $Z = X + Y$ is equivalent to:

$
\begin{align}
P_{Z}(z) &= \pmatrix{n+m \\ z} p^{z} (1-p)^{n+m-z} 
\end{align}
$

##### Code 2-i

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import comb
sns.set_theme(font_scale=1.0)
sns.set_style("white")
sns.set_style("ticks")
palette = sns.color_palette('deep')
```

```python
# let p = 0.3
p = 0.3 # Bernoulli parameter
n = 200
m = 50
z = np.arange(0, n+1)
X = binom(n, p)
Y = binom(m, p)
Z = binom(n+m, p)

i = 10000
Xrvs = X.rvs(i)
Yrvs = Y.rvs(i)
Zrvs = Xrvs + Yrvs

fig = plt.figure(figsize=(8,4))
plt.plot(z, Z.pmf(z), ms=5, color='deeppink', label='$P_{Z}(z)\sim Binomial(n+m,p)$')
bins = np.arange(0,n+1,2)
plt.hist(Zrvs,bins,density=True, label='$Z = X + Y$')
plt.xlabel('z')
plt.ylabel('Probability')
plt.grid()
plt.legend()
# plt.savefig('pdf_2.png', dpi=150, bbox_inches='tight', transparent=True)
plt.show()
```

Thus we can see from having plotted both $Z\sim B(n+m,p)$ and $Z = X + Y$, that indeed, if $X\sim B(n,p)$, $Y\sim B(m,p)$, and $Z = X + Y$,  then $Z\sim B(n+m,p)$

##### Math 2-ii

To show this relationship, we will simply follow the math, starting from the cumulative density function of $Z$ given $X$ (as mentioned in the prompt)

$
\begin{align}
F_{Z|X}(z|x) = P(Z \le z|x) &= P(Y + X \le z | x) \\
&= P(Y \le z - X | x) \\
&= P(Y \le z - x) \\
&= F_{Y}(z-x) \\
\end{align}
$

Differentiating:

$
\begin{align}
\frac{d}{dz}F_{Z|X}(z|x) &= \frac{d}{dz}F_Y(z-x) \\
\Rightarrow f_{Z|X}(z|x) &= \frac{d}{dy}F_Y(z-x)\frac{dy}{dz} \\
where\ Z &= Y+X \Rightarrow \frac{dz}{dy} = 1 \Rightarrow \frac{dy}{dz} = 1 \\ 
\\
\Rightarrow f_{Z|X}(z|x) &= f_Y(z-x)
\end{align}
$

Derive the conditional probability density function of $X$, given that $Z=z$, when $X \sim \mathcal{N} \left(0, 1\right)$ and $Y \sim \mathcal{N} \left(0, 1 \right)$.

##### Math 2-iii

To derive the conditional probability density function of $X$, given $Z=z$, we refer to Bayes' Theorem [3], which gives us that:

$
\begin{align}
f_{X|Z}(x|z) = \frac{f_{Z|X}(z|x) f_X(x)}{f_Z(z)}
\end{align}
$

Thus, subsituting in $f_{Z|X} \left( z | x \right) = f_{Y} \left( z - x \right)$, we have:

$
\begin{align}
f_{X|Z}(x|z) &= \frac{f_{Y} \left( z - x \right) f_X(x)}{f_Z(z)} \\
\end{align}
$

We already know that the pdf of a Gaussian Distribution is of the form [4]:

$$f_{X\sim \mathcal{N}(\mu,\sigma^2)}(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2 \right)$$

Thus we only need the distribution of the sum of the Gaussian random variables, i.e. $Z = X + Y$. From the identify for the Sum of Continuous variables [5], we have that:

$
\begin{align}
f_{Z}(z) &= \int_{-\infty}^{\infty} f_{X}(x) f_{Y}(z-x) dx
\end{align}
$

Plugging in for the distributions of $X\sim \mathcal{N}(0,1)$ and $Y\sim \mathcal{N}(0,1)$, we have:

$
\begin{align}
f_{Z}(z) &= \int_{-\infty}^{\infty} \left(\frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right) \right)
                                    \left(\frac{1}{\sqrt{2\pi}} \exp\left(-\frac{(z-x)^2}{2}\right) \right) dx \\
        &= \frac{1}{2\pi} \int_{-\infty}^{\infty} \exp\left( -\frac{1}{2}\left(x^2 + (z-x)^2\right) \right) dx
\end{align}
$

Evaluating this using MATLAB or a similar symbolic engine:

$
\begin{align}
f_{Z}(z) &= \int_{-\infty }^{\infty } \frac{\exp \left(-\frac{1}{2} \left(x^2 + (z-x)^2\right)\right)}{2 \pi } \, dx=\frac{e^{-\left.z^2\right/4}}{2 \sqrt{\pi }} = \frac{1}{\sqrt{2}\sqrt{2\pi}} \exp\left(-\frac{1}{2} \left(\frac{z^2}{2}\right)\right) \\
\Rightarrow Z &\sim \mathcal{N}(0,\sqrt{2})
\end{align}
$

Therefore, solving for the probability mass(density) function of x given z, we have:

$
\begin{align}
f_{X|Z}(x|z) &= \frac{f_{Y} \left( z - x \right) f_X(x)}{f_Z(z)} \\
             &= \frac{\left(\frac{1}{\sqrt{2\pi}} \exp\left(-\frac{(z-x)^2}{2}\right) \right)
                       \left(\frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right) \right)}
                       {\frac{1}{\sqrt{2} \sqrt{2\pi}}\exp \left(-\frac{1}{2}\left(\frac{z}{\sqrt{2}}\right)^2\right)} \\
             &= \frac{\sqrt{2}}{\sqrt{2\pi}} \exp \left(-\frac{1}{2}\left((z-x)^2 + x^2 - \frac{z^2}{2}\right)\right) \\
             &= \frac{1}{\frac{1}{\sqrt{2}}\sqrt{2\pi}} \exp \left(-\frac{1}{2}\left(\frac{z^2}{2} - 2xz + 2x^2\right)\right) \\
             &= \frac{1}{\frac{1}{\sqrt{2}}\sqrt{2\pi}} \exp \left(-\frac{1}{2}\left(\frac{x- \frac{z}{2}}{\frac{1}{\sqrt{2}}}\right)^2\right) \\ \\
\Rightarrow X|Z &\sim \mathcal{N}\left(\frac{z}{2},\frac{1}{\sqrt{2}}\right)\\
\end{align}
$

#### Part 2-iv

Prompt:

Compute the expectation of $1/\left(Z + 1 \right)$, when $X \sim Poisson \left( \lambda \right)$ and $Y \sim Poisson \left( \lambda \right)$. You may find [this Wikipedia page](https://en.wikipedia.org/wiki/Poisson_distribution) useful.

##### Math 2-iv

In order to compute the expectation of $1/(Z + 1)$, we first define a new random variable $W = \frac{1}{Z+1}$, and find the probability density function of $W$:

$\begin{align}
F_W(w) &= P(W \le w) = P\left(\frac{1}{Z+1} \le w\right) = P(1/w \le Z + 1) \\
 &= P(1/w - 1 \le Z) = 1 - P(Z \le 1/w - 1) \\ \\
 \Rightarrow \frac{d}{dw}(F_W(w)) &= \frac{d}{dw}\left(1 - P(Z \le 1/w - 1)\right) \\
 \Rightarrow P_W(w) &= -P_Z\left(\frac{1}{w} -1\right) \frac{dz}{dw},\\
 &\text{where}\ \frac{dz}{dw} = \frac{d}{dw}(1/w - 1) = -\frac{1}{w^2} \\ \\
 \Rightarrow P_W(w) &= P_Z\left(\frac{1}{w} -1\right) \frac{1}{w^2}
\end{align}$

Now that we have the distribution of W, we need to find the distribution of Z. We know that the formula for the Sum of Independent Discrete Random Variables, where $Z = X + Y$, is given by [1]:

$\begin{align}
P_Z(z) &= \sum_{x \in X} P_X(x) P_Y(z-x)
\end{align}$

Now, $X$ and $Y$ are Poisson random variables, and have a probability density function of the form [6]

$\begin{align}P(X=x) = \frac{\lambda^x e^{-\lambda}}{x!}\end{align}$

where $x \in \{0,1,2,...,\infty\}$ and $\lambda > 0$. Thus, 

$\begin{align}
P_Z(z) &= \sum_{x \in X} P_X(x) P_Y(z-x) \\
\Rightarrow P_Z(z) &= \sum_{x=0}^\infty \frac{\lambda^x e^{-\lambda}}{x!} \left(\frac{\lambda^{z-x} e^{-\lambda}}{(z-x)!} \right) \Rightarrow \text{MATLAB} \\
                   &= \frac{2^z \lambda^z e^{-2\lambda}}{z!}
\end{align}$

Where $z \in \{0,1,2,...,\infty\}$. Note that this is the same as $Z \sim Poisson(2\lambda)$, which is given in the [wikipedia article](https://en.wikipedia.org/wiki/Poisson_distribution) above.

Thus, the pdf of $W$ is:

$\begin{align}
P_W(w) = \frac{{\mathrm{e}}^{-2\,\lambda } \,{{\left(2\,\lambda \right)}}^{\frac{1}{w}-1} }{w^2 \,{\left(\frac{1}{w}-1\right)}!},\ \text{for}\ w \in \{1, 1/2, 1/3, ...\}
\end{align}$

To check whether this is a real pdf or not, we compute the cumulative pdf for all values of w and check whether it is equal to 1:

$\begin{align}
\sum_{w\in \mathcal{W}} P_W(w) &= \sum_{w\in \mathcal{W}} \frac{{\mathrm{e}}^{-2\,\lambda } \,{{\left(2\,\lambda \right)}}^{\frac{1}{w}-1} }{w^2 \,{\left(\frac{1}{w}-1\right)}!} = \sum_{n=1}^{\infty}\frac{n^2 \,{\mathrm{e}}^{-2\,\lambda } \,{{\left(2\,\lambda \right)}}^{n-1} }{{\left(n-1\right)}!},\ \text{where we have substituted}\ w = \frac{1}{n}\ \text{s.t.}\ n \in \{1,2,3,...,\infty \} \\ \\
\Rightarrow \text{MATLAB} \\
\Rightarrow \sum_{w\in \mathcal{W}} P_W(w) &= 4\lambda^2 + 6\lambda + 1 \ne 1
\end{align}$

And we note that the cumulative distribution function of $W$ is not always equal to one! 

Since $\sum_{w\in \mathcal{W}} P_W(w) \ne 1$, we define:

$\begin{align}\hat{P}_W(w) = \frac{P_W(w)}{\sum_{w\in \mathcal{W}} P_W(w)}\end{align}\ 
\text{such that}\ \sum_{w\in \mathcal{W}} \hat{P}_W(w) = 1
$

Now, we can compute the expectation of $W$:

$\begin{align}
&\mathbb{E}_W [w] = \sum_{w\in \mathcal{W}} w P_W(w) \\
\Rightarrow &\text{MATLAB} \\
\Rightarrow &\mathbb{E}_W [w] = \frac{2\lambda + 1}{4\lambda^2 + 6\lambda + 1}
\end{align}$

## Problem 3

A new test for a disease has been produced by a company. Trials revealed that if a test user has the disease, the test is positive with $95\%$ probability. Similarly, if the user does not have the disease, the test is negative with $92 \%$ probability. Approximately $1\%$ of the population is suspected to have this disease. Compute the probabilities below:

(i) The probability that a randomly chosen individual will test positive. [4]

(ii) The probability that a randomly chosen individual who tests positive will have the disease. [4]

Total points: [8] 

### Solution Submission

#### Math 3

Let the following nomenclature hold:

$
\begin{align*}
T+ &= \textrm{positive test} \\
T- &= \textrm{negative test} \\
D &= \textrm{test user with disease} \\
\bar{D} &= \textrm{test user absent the disease}
\end{align*}
$

Then, from the prompt we are given:

$
\begin{align*}
P(T+|D) &= 0.95 \\
P(T-|\bar{D}) &= 0.92 \\
P(D) &= 0.01 \\
P(\bar{D}) &= 0.99
\end{align*}
$

----

To answer **part (i)**, we are looking for $P(T+)$. By the Sum Rule [1] and Conditional Probablity [2], we are given:

$
\begin{align*}
P(T+) &= P(T+,D) + P(T+,\bar{D}) \\
&= P(T+|D)P(D) + P(T+|\bar{D})P(\bar{D})
\end{align*}
$

And by the Complement Rule [3] we can find:

$\begin{align*}
P(T+|\bar{D}) = 1 - P(T-|\bar{D}) = 0.08
\end{align*}
$

---

To answer **part (ii)**, we are looking for $P(D|T+)$. By Bayes' Thm [4], we have:

$
\begin{align*}
P(D|T+) = \frac{P(T+|D)P(D)}{P(T+)}
\end{align*}
$


#### Code + Answer 3

```python
from IPython.display import display, Latex
```

```python
P_t_plus_given_d = 0.95
P_t_min_given_dbar = 0.92
P_d = 0.01
P_dbar = 0.99

# Part (i)
P_t_plus_given_dbar = 1 - P_t_min_given_dbar
P_t_plus = P_t_plus_given_d*P_d + P_t_plus_given_dbar*P_dbar

display(Latex(f"$P(T+)={P_t_plus:.5f}$"))

# Part (ii)
P_d_given_t_plus = P_t_plus_given_d*P_d/P_t_plus

display(Latex(f"$P(D|T+)={P_d_given_t_plus:.5f}$"))
```

#### Part 4-i

For part 4-(i) I will select a logarithmic function in combination with an exponential function such that the basis becomes

$X = \begin{bmatrix}1 & \log(x) & \exp(x)\end{bmatrix}$

```python
# Data
df = pd.read_csv('data100m.csv')
df.columns=['Year', 'Time']
N = df.shape[0]

# defining x and y (i.e. "t") variables
max_year, min_year = df['Year'].values.max() , df['Year'].values.min()
x_offset = 0.1 # prevents division by 0 in the basis log function
x = (df['Year'].values.reshape(N,1) - min_year)/(max_year - min_year) + x_offset
# x = df['Year'].values.reshape(N,1)
t = df['Time'].values.reshape(N,1)

# Basis
X_func = lambda u : np.hstack([np.ones((u.shape[0],1)), np.log(u), np.e**u])
X = X_func(x)
```

Now, generating the multivariate normal Prior and calculating the Posterior estimates of the best model parameters for the given data and basis.

Note that we have to generate a Sigma_0 matrix with a low condition number since otherwie it becomes numerically intractable for computing the inverse of Sigma_0, which is required for computer Sigma_w and mu_w of the posterior.

```python
# Prior
sigma_hat = 0.2  # variance of the likelihood (i.e. assumed observation noise)

# from: Probablistic Machine Learning - Introduction, [2]
# https://github.com/probml/pyprobml/blob/master/notebooks/book1/04/shrinkcov_plots.ipynb
# ----------------
# A function that generates a random covariance matrix (and its inverse) with the given condition number
# and with the given first axis a (unnormalised). Returns (Sigma, Lambda).
def covcond(cond_number, a):
    # Use the Householder trick to generate a basis Z whose first axis points in the direction of a.
    a = np.copy(a)
    a[0] = a[0] + np.sign(a[1]) * np.linalg.norm(a)
    Z = np.eye(len(a)) - 2 * np.matmul(a.reshape(-1, 1), a.reshape(1, -1)) / np.linalg.norm(a) ** 2

    e = np.flip(np.sort(1.0 / np.linspace(cond_number, 1, len(a))))
    Sigma = np.matmul(np.matmul(Z, np.diag(e)), Z.transpose())
    Lambda = np.matmul(np.matmul(Z, np.diag(1 / e)), Z.transpose())

    return (Sigma, Lambda)
# ----------------

cond_number = 10
a = 0 + np.random.randn(m, 1)*sigma_hat  # mean 0, standard deviation of sigma_hat
Sigma_0, Lambda = covcond(cond_number, a)

inv_Sigma_0 = np.linalg.inv(Sigma_0)
prior = multivariate_normal(mu_0.flatten(), Sigma_0)

# Condition Number of Sigma_0: [3] - reference to Probabilistic Machine Learning - Introduction, pg 236, section 7.1.4.4
display(Latex(r"Condition Number of $\Sigma_0$={:.2f}".format(np.linalg.cond(Sigma_0))))

fig = plt.figure(figsize=(5,4))
random_samples = 200
plt.plot(xi, Xg @ prior.rvs(random_samples).T, alpha=0.7)
plt.xlabel('Year')
plt.title('model using prior samples')
plt.ylabel('Time (seconds)')
plt.show()
```

```python
# Posterior Parameters
sigma_hat = 0.3  # variance of the likelihood (i.e. assumed observation noise)
Sigma_w = np.linalg.inv(1./(sigma_hat**2) * (X.T @ X) + inv_Sigma_0)
mu_w = Sigma_w @ (1./(sigma_hat**2) * (X.T @ t) + (inv_Sigma_0 @ mu_0) )

# Posterior
posterior = multivariate_normal(mu_w.flatten(), Sigma_w)

display(Latex('Posterior weight means, $\mu_w = $'))
display(posterior.mean)

# Condition Number of Sigma_w:
display(Latex(r"Condition Number of $\Sigma_w$={:.2f}".format(np.linalg.cond(Sigma_w))))

# Variance of w:
display(Latex('Variances of the basis weights:'))
display(np.diagonal(Sigma_w))

fig = plt.figure(figsize=(7,5))
random_samples = 200
plt.plot(xi, Xg @ posterior.rvs(random_samples).T, zorder=-1, alpha=0.4)
a, = plt.plot(df['Year'].values, df['Time'].values, 'o', color='dodgerblue', \
              label='Data', markeredgecolor='k', lw=1, ms=10, zorder=1)
b, = plt.plot(xi, Xg @ posterior.mean.T, linewidth=1.5, color='red')
plt.xlabel('Year')
plt.title('model using posterior samples')
plt.legend([a, b], ['Data', 'Posterior Mean'], framealpha=0.2)
plt.ylabel('Time (seconds)')
plt.show()
```
