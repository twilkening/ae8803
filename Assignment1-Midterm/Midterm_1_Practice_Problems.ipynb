### Problem 1

```python
X = 2 + 2*np.random.rand(9000)

w = np.linspace(8, 64, 100)
fz = lambda u : 1/6*(1/u**(2/3))

fig = plt.figure(figsize=(6,3))
plt.hist(X**3,64-8, density=True, color='orangered',label=r'$X^{3}$')
plt.plot(w,fz(w),'-', color='dodgerblue', label=r'$f_{W}(w))$')
plt.ylabel(r'$f_{W}(w))$')
plt.xlabel('w')
plt.legend()
plt.show()
```

### Problem 3

3. **Function of a Random Variable**: Assume $ X $ is a random variable with a normal distribution, mean $ \mu = 3 $, and variance $ \sigma^2 = 9 $. Calculate the probability density function of $ Y = e^X $.


```python
X = np.random.normal(3,9,18000)
Y = np.exp(X)

y = np.linspace(1, 100, 200)
fy = lambda u : 1/(3*(2*np.pi)**(1/2)) * np.exp(-1/(2*9) * (np.log(u) - 3)**(2)) * 1/u

fig = plt.figure(figsize=(6,3))
plt.hist(Y[Y < 500],100, density=True, color='orangered',label=r'$e^X$')
plt.plot(y,fy(y),'-', color='dodgerblue', label=r'$f_{Y}(y))$')
plt.ylabel(r'$f_{Y}(y))$')
plt.xlabel('y')
plt.legend()
plt.show()
```
